```python
import pandas as pd

person_1=pd.Series({
                     'age':15,
                     'place':'Manipal'})
person_2=pd.Series({
                     'age':16,
                     'place':'udupi'})

person_3=pd.Series({
                     'age':20,
                     'place':'Mangalore'})
person_4=pd.Series({
                     'age':20,
                     'place':'Bangalore'})
person_5=pd.Series({
                     'age':17,
                     'place':'Kundapur'})

df=pd.DataFrame([person_1,person_2,person_3,person_4,person_5],index=['Ravi','sanju','cs','Mani','Akshay'])
df.head()
```

```python
df.insert(2, "cost", [22,214.0,22,16,20],True) 
  
df
```

```python
person_6=pd.Series({
                     'age':21,
                     'place':'Udupi',
                    'cost':30})
person_7=pd.Series({
                     'age':17,
                     'place':'Shivamogga',
                      'cost':40})
df1=pd.DataFrame([person_6,person_7],index=['Raju','Bheem'])

df=df.append(df1) 
df
```

```python
copy_df=df.copy()
copy_df
```

```python
copy_df=copy_df.drop('Ravi')
copy_df=copy_df.drop('sanju')
copy_df
```
